**Statistical Analysis of Dataframe**

```python
# Get statistical analysis of your dataframe, such as count, mean, min, max
df.describe()
```

Sorting data:

```python
df.sort_values('Feedback')
```

Sorting in descending order:

```python
df.sort_values('Avg Rating', ascending = False)
```

Sort using two columns:

```python
df.sort_values(['Destination', 'Avg Rating'], ascending = [True, False])
# This will sort the data in the following order:- 
# First it'll sort on the basis of 'Destination' column in ascending format
# Then it'll sort on the basis of 'Avg Ratings' column in descending format
```
